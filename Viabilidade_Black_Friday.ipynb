# Vendas na Black Friday

A Black Friday é uma tradição de compras que teve início nos Estados Unidos e ocorre anualmente na sexta-feira seguinte ao Dia de Ação de Graças, que é celebrado na quarta quinta-feira de novembro. A data marca o início da temporada de compras natalícias.

A ideia por trás da Black Friday é oferecer grandes descontos e promoções em produtos e serviços, incentivando os consumidores a fazerem compras, muitas vezes antecipando suas compras de Natal. As lojas físicas e online participam ativamente da Black Friday, oferecendo descontos significativos em uma variedade de produtos, desde eletrônicos e eletrodomésticos até roupas, brinquedos e muito mais.

A expressão "Black Friday" tem origens no final do século XIX e inicialmente era associada a crises financeiras e congestionamentos nas ruas causados pelo aumento do tráfego devido às compras de Natal. No contexto do comércio, o termo "black" passou a ser associado ao lucro, indicando que as lojas passariam de números vermelhos (prejuízo) para números pretos (lucro) devido ao aumento das vendas durante esse período.

A Black Friday tornou-se um fenômeno global, com muitos países adotando a tradição e os consumidores aguardando ansiosamente por ofertas especiais nesse dia. No entanto, a Black Friday também é objeto de críticas devido a situações como tumultos em lojas físicas, pressão sobre os trabalhadores e a possibilidade de promoções enganosas. Algumas empresas estenderam a Black Friday para o que é chamado de "Cyber Monday", oferecendo descontos especiais em compras online na segunda-feira seguinte.

**Valores Históricos

-Este valores podem ser modificados para atender a realidade do negócio

```python
 # Unidades vendidas na semana anterior
volume_total_anterior = 1000.00

#Custo unitário de compra no fornecedor
custo_unitario = 100  

 # o preço bruto no ponto de venda
preco_venda_atual = 120 

#Lucro na semana anterior
lucro_total_inicial = volume_total_anterior * (preco_venda_atual - custo_unitario)
```

**Premissas

-Este valores podem ser modificados para atender a realidade do negócio

```python
#% de desconto por unidade vendida
desconto = 0.10 # em cada unidade vendida

#Além do volume de vendas o comerciante espera aumentar o Lucro final do negócio
aumento_lucro_percentual = 0.20 
```

**Cálculos e Viabilidade

```python

#Novo preço de venda do produto considerando o desconto da black friday
Novo_preço = preco_venda_atual-(preco_venda_atual* desconto)

#obviamente o desconto fornecido implica em perda de margem a cada unidade vendida
Perda_desc = preco_venda_atual-(Novo_preço)

#Para recuperar a perda no lucro total do negócio eu preciso vender mais unidades apenas para recuperar o volume //
#nominal do lucro da semana anaterior
unidades_adicionais_vendidas = (preco_venda_atual * desconto * volume_total_anterior) / preco_venda_atual

#Mas eu espero com o aumento das vendas recuperar a lucratividade ganhando em escala
Novo_lucro= (lucro_total_inicial*aumento_lucro_percentual)+lucro_total_inicial

#Para atingir o o faturamento e aumentar a base nominal de lucro atual é preciso vender além da semana anterior
Unidades= (Novo_lucro / Novo_preço) + unidades_adicionais_vendidas+volume_total_anterior

# Além do volume normal de vendas eu preciso vender aproximadamente 30% a mais em relação as semanas ateriores
Esforço = Unidades - volume_total_anterior
```

**Resumo descritivo

```python
print(f"Perda unitária: R$ {Perda_desc: .2f}")

print(f"Novo preço de venda: R$ {Novo_preço:.2f}")

print (f"Qtd adicional para recuperar o lucro inicial: {unidades_adicionais_vendidas:.0f} unidades")

print(f"Novo lucro esperado com 10% de aumento: R${Novo_lucro: .02f}")

print(f"Unidades necessária pra atingir o novo lucro: {Unidades: .0f} unidades")

print(f"Esforço adicional de vendas na Black Friday: {Esforço: .0f} unidades ")
```

**Visualização Gráfica

```python
import matplotlib.pyplot as plt

# Dados para o gráfico
lucros = [lucro_total_inicial, Novo_lucro]
labels = ['Semana anterior', 'Black friday']

# Plotando o gráfico
plt.bar(labels, lucros, color=['blue', 'green'])
plt.xlabel('Situação')
plt.ylabel('Lucro')
plt.title('Comparação - Lucro esperado')
plt.show()
```
